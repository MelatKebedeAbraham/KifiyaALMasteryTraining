```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Load the dataset
data = pd.read_csv('C:/Users/user/Desktop/Kifiya AI Mastery Traning/Data/Processed/merged_data.csv')

# Display the first few rows
data.head()
```

```python
summary_stats = data.describe()
print(summary_stats)
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data.drop('Comments', axis=1, inplace=True)
```

```python
import seaborn as sns

sns.boxplot(data=data[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']])
plt.show()
```

```python
data['time'] = pd.to_datetime(data['time'])
data.set_index('time', inplace=True)

plt.figure(figsize=(14, 7))
plt.plot(data['GHI'], label='GHI')
plt.plot(data['DNI'], label='DNI')
plt.plot(data['DHI'], label='DHI')
plt.plot(data['Tamb'], label='Tamb')
plt.legend()
plt.show()
```

```python
cleaned_data = data[data['Cleaning'] == 1]
not_cleaned_data = data[data['Cleaning'] == 0]

plt.figure(figsize=(14, 7))
plt.plot(not_cleaned_data['ModA'], label='ModA - Before Cleaning')
plt.plot(cleaned_data['ModA'], label='ModA - After Cleaning')
plt.legend()
plt.show()
```

```python
correlation_matrix = data[['GHI', 'DNI', 'DHI', 'TModA', 'TModB', 'WS', 'WSgust', 'WD']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(projection='polar')
plt.scatter(data['WD'] * np.pi / 180, data['WS'])
plt.title('Wind Speed and Direction')
plt.show()
```

```python
sns.scatterplot(x='RH', y='Tamb', data=data)
plt.title('Relative Humidity vs Temperature')
plt.show()
```

```python
data[['GHI', 'DNI', 'DHI', 'WS', 'Tamb']].hist(bins=30, figsize=(10, 8))
plt.show()
```

```python
from scipy.stats import zscore

data['z_score'] = zscore(data['GHI'])
outliers = data[data['z_score'].abs() > 3]
print(outliers)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['GHI'], data['Tamb'], s=data['WS']*10, c=data['RH'], alpha=0.5)
plt.xlabel('GHI')
plt.ylabel('Tamb')
plt.title('GHI vs Tamb vs WS with Bubble Size representing RH')
plt.colorbar(label='Relative Humidity')
plt.show()
```

```python
# Drop rows where Comments are null if Comments is necessary
data.dropna(subset=['Comments'], inplace=True)

# Replace missing values in GHI, DNI, and DHI with mean or median
data['GHI'].fillna(data['GHI'].mean(), inplace=True)
data['DNI'].fillna(data['DNI'].mean(), inplace=True)
data['DHI'].fillna(data['DHI'].mean(), inplace=True)

# Handling negative values where only positive should exist
data[['GHI', 'DNI', 'DHI']] = data[['GHI', 'DNI', 'DHI']].clip(lower=0)
```
